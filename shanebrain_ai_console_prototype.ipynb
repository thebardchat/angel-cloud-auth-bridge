```
import json
import pandas as pd
import math # Import math to check for nan

# Function to safely parse the string representation of the details dictionary
def parse_details(details_str):
    # Handle None or string 'nan' explicitly
    if details_str is None or (isinstance(details_str, str) and details_str.lower() == 'nan'):
        return {}
    try:
        # The string representation includes single quotes, convert to double quotes for json.loads
        details_str = details_str.replace("'", '"')
        # Safely load the JSON string
        return json.loads(details_str)
    except (json.JSONDecodeError, TypeError): # Catch other parsing errors and non-string inputs
        # Return an empty dictionary if parsing fails or input is invalid
        return {}

# Apply the parsing function to the 'details' column
audit_log_df['parsed_details'] = audit_log_df['details'].apply(parse_details)

# Now you can access elements within the parsed_details column
print("Parsed Details from Audit Log:")
for index, row in audit_log_df.iterrows():
    print(f"Action: {row['action']}, Details: {row['parsed_details']}")

# You can also create new columns from the parsed details
audit_log_df['trip_minutes'] = audit_log_df['parsed_details'].apply(lambda x: x.get('trip_minutes'))
audit_log_df['complexity_factor'] = audit_log_df['parsed_details'].apply(lambda x: x.get('complexity_factor'))

# Display the DataFrame with the new parsed columns
display(audit_log_df)
```

```
import datetime
import hashlib
import time
import math

# --- Simulated Data Structures ---
# Simulate Firestore data for dispatcher logic
simulated_firestore = {
    'artifacts': {
        'default-app-id': {
            'users': {
                'default-user': {
                    'dispatcher_logic': {
                        'HaulRateCalculator': {
                            'logicName': 'HaulRateCalculator',
                            'baseRatePerHour': 130,
                            'minutesPerCycle': 60,
                            'tonnageDivisor': 25,
                            'minimumRate': 6,
                            'roundingRule': 'Nearest half dollar ($0.50) up, minimum $6.00.',
                            'formulaTemplate': '(\\frac{\\text{baseRatePerHour}}{\\text{minutesPerCycle}} \\times \\text{roundTripMinutes}) \\div \\text{tonnageDivisor}'
                        }
                    }
                }
            }
        }
    }
}

# Simulate Pulsar Sentinel audit log
pulsar_audit_log = []

# --- Core Functionalities (Python Simulation) ---

def get_dispatcher_logic(app_id: str, user_id: str) -> dict | None:
  """Simulates fetching dispatcher logic from Firestore."""
  try:
    return simulated_firestore['artifacts'][app_id]['users'][user_id]['dispatcher_logic']['HaulRateCalculator']
  except KeyError:
    return None

def calculate_haul_rate(dispatcher_logic: dict, trip_minutes: int, complexity_factor: float = 1.0) -> float | None:
  """Calculates the haul rate based on dispatcher logic, with an added complexity factor."""
  if not dispatcher_logic or trip_minutes is None or trip_minutes <= 0:
    return None

  # Apply complexity factor to minutes per cycle (e.g., rough terrain increases time)
  adjusted_minutes_per_cycle = dispatcher_logic['minutesPerCycle'] * complexity_factor

  # 1. Core Calculation
  rate_per_minute = dispatcher_logic['baseRatePerHour'] / adjusted_minutes_per_cycle
  raw_rate = (rate_per_minute * trip_minutes) / dispatcher_logic['tonnageDivisor']

  # 2. Apply Custom Rounding Logic (Nearest half dollar up, min 6)
  final_rate = max(raw_rate, dispatcher_logic['minimumRate'])
  final_rate = math.ceil(final_rate * 2) / 2

  return final_rate

def sha256_hash(data: str) -> str:
    """Generates a SHA-256 hash for the given data."""
    return hashlib.sha256(data.encode('utf-8')).hexdigest()

def anchor_transaction(user_id: str, trip_minutes: int, calculated_rate: float, complexity_factor: float = 1.0):
    """Simulates creating and anchoring a financial transaction log, including complexity."""
    action_description = f"LogiBot Quote: ${calculated_rate:.2f}/ton submitted to client (Complexity: {complexity_factor:.1f}x)."
    data_to_hash = f"{action_description}|{datetime.datetime.now().isoformat()}|{user_id}|{trip_minutes}|{complexity_factor}"
    hash_value = sha256_hash(data_to_hash)

    new_log = {
        'timestamp': datetime.datetime.now().strftime('%H:%M:%S'),
        'action': action_description,
        'value': f"{calculated_rate:.2f}",
        'hash': hash_value,
        'status': 'PENDING',
        'details': {'trip_minutes': trip_minutes, 'complexity_factor': complexity_factor}
    }

    pulsar_audit_log.append(new_log)
    print(f"Transaction logged: {new_log}")

    # Simulate Asynchronous Pulsar Blockchain Anchoring (5-second delay)
    # In a real application, this would be a background process
    time.sleep(5)
    for log in pulsar_audit_log:
        # Find the log by hash and complexity_factor (since hash is based on details)
        if log['hash'] == hash_value and log.get('details', {}).get('complexity_factor') == complexity_factor:
            log['status'] = 'ANCHORED'
            print(f"Transaction anchored: {log}")
            break


def download_legacy_stick(user_id: str):
    """Simulates the Legacy AI Stick download and anchoring, including file structure details."""
    base_cost = 499.00
    profile_data = f"Shanebrain Legacy AI Profile Export for User {user_id} - Base Cost: ${base_cost:.2f}"
    data_to_hash = f"{profile_data}|{datetime.datetime.now().isoformat()}|LEGACY_EXPORT_T3"
    hash_value = sha256_hash(data_to_hash)

    print(f"Legacy AI Stick generated! Cost: ${base_cost:.2f}. Pulsar Anchor: {hash_value[:10]}... (Tier 3)")

    print("\nSimulated Legacy Stick File Structure:")
    print("  - CORE_PROFILE.SHANE: Contains encrypted, read-only core AI profile and licensed IP. Secured by Pulsar Sentinel (Tier 3).")
    print("  - USER_STORAGE.DAT: Encrypted read/write storage for user's personal data, worlds, and games. (Upgradable capacity/speed).")


    # Also log this final, critical action to the audit log
    new_log = {
      'timestamp': datetime.datetime.now().strftime('%H:%M:%S'),
      'action': 'Legacy AI Profile Export Initiated (T3).',
      'value': f"{base_cost:.2f}",
      'hash': hash_value,
      'status': 'PENDING',
    }

    pulsar_audit_log.append(new_log)
    print(f"Legacy export logged: {new_log}")

    # Simulate final anchoring for the Legacy Stick export
    time.sleep(5)
    for log in pulsar_audit_log:
        if log['hash'] == hash_value:
            log['status'] = 'ANCHORED'
            print(f"Legacy export anchored: {log}")
            break
```
